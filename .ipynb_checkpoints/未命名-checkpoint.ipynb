```python
## these part is used to do some data preprocessing and transfer the txt to json data format
import numpy as np
import json

# for buyer information
# data import
buyer_file = np.loadtxt('buyer_basic_info.txt')
buyer_file = buyer_file.astype(np.int32)
print(buyer_file.shape)

quantity_file = np.loadtxt('buyer_historical_category15_quantity.txt')
money_file = np.loadtxt('buyer_historical_category15_money.txt')
print(quantity_file.shape)
print(money_file.shape)

# the dic index with buyer_id
# basic_info: 0 for register_time
#                 1 for seller_level
#                 2 for buyer_level
#                 3 for age
#                 4 for gender
#                 all above default with -1
# quantity: n means the consumption quantity in the nth category
# money: ditto
# average_money: ditto
buyer_basic_info = {}
for buyer in buyer_file:
    buyer_basic_info[str(buyer[0])] = {}
    buyer_basic_info[str(buyer[0])]['basic_info'] = buyer[0: 6].tolist()
    
for quantity, money in zip(quantity_file, money_file):
    quantity = quantity.astype(np.int32)
    buyer_basic_info[str(int(quantity[0]))]['quantity'] = quantity.tolist()
    buyer_basic_info[str(int(quantity[0]))]['money'] = money.tolist()
    average_money = []
    for a, b in zip(quantity, money):
        if a != 0:
            average_money.append(b/a)
        else:
            average_money.append(-1)
    buyer_basic_info[str(int(quantity[0]))]['average_money'] = average_money

# write data
with open('./buyer_basic_info.json', 'w') as f:
    json.dump(buyer_basic_info, f, indent=4)
    print("storing the buyer data finished!")

# for product information
# import data
key_product = np.loadtxt('key_product_IDs.txt')
product_feature = np.loadtxt('product_features.txt')
trade_info = np.loadtxt('trade_info_training.txt')
print(key_product.shape)
print(product_feature.shape)
print(trade_info.shape)

# the dic index with the product id
# basic_info: 0 for feature 1
#                 1 for feature 2
#                 2 for original price
# trade: store every trade information
product_info = {}
for product in product_feature:
    if product[0] in key_product:
        product_info[str(int(product[0]))] = {}
        product_info[str(int(product[0]))]['basic_info'] = product[0: 4].tolist()
        product_info[str(int(product[0]))]['trade']  = []


for trade in trade_info:
    product_info[str(int(trade[0]))]['trade'].append(trade.tolist())

for key in product_info:
    product_info[key]['trade'] = sorted(product_info[key]['trade'], key=lambda x:x[2])

# write data
with open('./product_info.json', 'w') as f:
    json.dump(product_info, f, indent=4)
    print("storing the product data finished!")
```

```python
## data redundancy detect, data reduction and correlation analysis
import numpy as np
import json
import matplotlib.pyplot as plt
with open('product_info.json', 'r') as f:
    product_info = json.load(f)

key_product_basic_info = np.zeros([100, 4])
key_product_distribute = np.zeros([101, 120])
product_distribute_set = np.loadtxt('product_distribution_training_set.txt')
delta_product = np.zeros([100, 119])

# extract the key product basic information
# sum the trade information to calculate the product distribution
# and compare with the data in 'product_distribution_training_set.txt'
row = 0
for key in product_info:
    for i in range(4):
        key_product_basic_info[row][i] = product_info[key]['basic_info'][i]
    key_product_distribute[row][0] = int(key)
    for trade in product_info[key]['trade']:
        key_product_distribute[row][int(trade[2]) + 1] += int(trade[3])
    row += 1
for i in range(120):
    key_product_distribute[100][i] = np.sum(key_product_distribute[:, i])
key_product_basic_info = sorted(key_product_basic_info, key=lambda x:x[0])

key_product_distribute = sorted(key_product_distribute, key=lambda x:x[0])

for i in range(100):
    for j in range(119):
        delta_product[i][j] = key_product_distribute[i][j] - product_distribute_set[i][j]

attribute1 = [a[1] for a in key_product_basic_info]
print('the attribute1:')
print(attribute1)
attribute2 = [a[2] for a in key_product_basic_info]
print('the attribute2:')
print(attribute2)
print('the error of the list:')
print(sum(delta_product))
key_product_distribute = np.array(key_product_distribute).astype(np.int32)
print(key_product_distribute.shape)
np.savetxt('key_product_distribute.txt', key_product_distribute, fmt="%d")
reduced_product_info = np.zeros([100, 2])
for i in range(100):
    reduced_product_info[i][0] = key_product_basic_info[i][1]
    reduced_product_info[i][1] = key_product_basic_info[i][3]
np.savetxt('reduced_product_info.txt', reduced_product_info,  fmt="%d")
print("data store success!")

# correlation analysis
correlation_mat = np.zeros([100, 4])
for i in range(100):
    correlation_mat[i][0] = key_product_basic_info[i][0] # product_id
    correlation_mat[i][1] = key_product_basic_info[i][1]  # attribute1
    correlation_mat[i][2] = key_product_basic_info[i][3] # price
    correlation_mat[i][3] = np.sum(key_product_distribute[i][1:120]) # sale quantity
plt.plot(correlation_mat[:, 2], correlation_mat[:, 3], '*')
plt.show()
plt.plot(correlation_mat[:, 1], correlation_mat[:, 3], '*')
plt.show()
plt.plot(correlation_mat[:, 1], correlation_mat[:, 2], '*')
plt.show()
```

```python
# time series analysis
# the training error is too much
# this method failed
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
import numpy as np
from statsmodels.tsa.arima_model import ARIMA

names = ['product_id']
for i in range(119):
    names.append(str(i)+ 'th')
data = pd.read_csv('key_product_distribute.txt', names=names, delim_whitespace=True, index_col='product_id')
product_sum = data.loc[28575]
product_sum.plot(color='blue', label='Raw Data')
rol_mean = product_sum.rolling(window=7, axis=0).mean()
rol_mean.plot(color='red', label='Rolling Mean')
sum_diff = product_sum - product_sum.shift()
sum_diff.dropna(inplace=True)
plt.show()
sum_diff.plot(color='red')
sum_diff2 = sum_diff - sum_diff.shift()
sum_diff2.dropna(inplace=True)
plt.show()
sum_diff2.plot(color='red')

decomposition = seasonal_decompose(product_sum, freq=14)  
fig = plt.figure()  
fig = decomposition.plot()  
fig.set_size_inches(15, 8)

from statsmodels.tsa.stattools import adfuller   #Dickey-Fuller test

def test_stationarity(timeseries):
    #Determing rolling statistics
    rolmean = pd.Series(timeseries).rolling(window=12).mean()
    rolstd = pd.Series(timeseries).rolling(window=12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)

    #Perform Dickey-Fuller test:
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)

test_stationarity(sum_diff2)

from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(sum_diff2, nlags=20)
lag_pacf = pacf(sum_diff2, nlags=20, method='ols')

#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(rol_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(rol_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(rol_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(rol_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
plt.show()

product = [a for a in product_sum]
model = ARIMA(product, order=(1, 2, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(sum_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-sum_diff2)**2))
plt.show()

model = ARIMA(product, order=(0, 2, 1))  
results_MA = model.fit(disp=-1)  
plt.plot(sum_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-sum_diff2)**2))
plt.show()

model = ARIMA(product, order=(1, 2, 1))  
results_ARIMA = model.fit()  
plt.plot(sum_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-sum_diff2)**2))
```
